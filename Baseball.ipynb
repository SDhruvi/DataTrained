```python
#Removing the outliers
from scipy.stats import zscore
z_score=abs(zscore(data))
print(data.shape)
```

```python
df=data.loc[(z_score<3).all(axis=1)]
print(df.shape)
```

```python
x=df.iloc[:,1:17]
x.shape
```

```python
y=df.iloc[:,0:1]
y.shape
```

```python
#linear regression
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
```

```python
max_r_score=0
for r_state in range(42,100):
    x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=.30,random_state=r_state)
    lr=linear_model.LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print(r_state,"is: ",r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state

print()
print(final_r_state,"is ",max_r_score)
```

```python
#applying cross validation techniques:
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
cross_val_score(linear_model.LinearRegression(),x,y,cv=5,scoring="r2")
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
#lasso

isreh=Lasso()
parameters={"alpha":[0.001,0.01,0.1,1]}
grid=GridSearchCV(isreh,parameters,cv=10)
grid.fit(x,y)
grid.best_params_
```

```python
max_r_score=0
for r_state in range(42,100):
    x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=.30,random_state=r_state)
    lasso=Lasso(alpha=0.01)
    lasso.fit(x_train,y_train)
    y_pred=lasso.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print(r_state,"is: ",r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print()
print()
print(final_r_state,"is ",max_r_score)
```

```python
#cross validation
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
cross_val_score(linear_model.LassoCV(),x,y,cv=5,scoring="r2")
```
